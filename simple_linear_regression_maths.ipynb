```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=([1,2,4,3,5,6,4])
Y=([1,3,3,2,5,4,4])
```

```python
print(X)
print(Y)
```

```python
data=pd.DataFrame({'X':X,'Y':Y})
data
```

```python
x_mean=data['X'].mean().round(2)
x_mean
```

```python
y_mean=data['Y'].mean().round(2)
y_mean
```

```python
data['X-X_dash']=data['X']-x_mean
data['Y-Y_dash']=data['Y']-y_mean
data['X-X_dash^2']=data['X-X_dash']**2
data['X-X_dash * Y-Y_dash']=data['X-X_dash']*data['Y-Y_dash']
data
```

```python
variance=data['X-X_dash^2'].sum()/6
variance
```

```python
covariance=data['X-X_dash * Y-Y_dash'].sum()/6
covariance
```

```python
b=covariance/variance
b
```

```python
a=y_mean-x_mean*b
a
```

```python
x='x'
```

```python
equation=(f"{a}+{b}x")
print(equation)
```

```python
y_pred=a+b*data['X']
```

```python
#print(f"y={a:.2f}+{b:.2f}x")
```

```python
plt.scatter(data['X'],data['Y'])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.plot(data['X'],y_pred)
plt.xlabel("X")
plt.ylabel("Y prediction")
plt.show()
```
